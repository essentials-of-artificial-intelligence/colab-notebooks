# Code Cell 2: Functions, `*args`, `**kwargs`, and Decorators

## Explanation

This cell introduces multiple important Python concepts at once.

A function is a reusable block of code. In AI engineering, functions are used for loading data, training models, evaluating performance, saving outputs, and preprocessing text or images.

`*args` allows a function to accept any number of positional arguments. In this example, the model layers are passed as positional arguments.

`**kwargs` allows a function to accept any number of keyword arguments. In this example, learning rate and epochs are passed as keyword arguments.

A decorator is a function that wraps another function and adds extra behavior. Here, the decorator measures how long the model compilation function takes.

This is useful in AI pipelines because training, preprocessing, and inference time matter.

````python
import time

def time_epoch(func):
    """Decorator to measure execution time."""
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        print(f"[LOG] Execution took {time.time() - start:.4f}s")
        return result
    return wrapper

@time_epoch
def compile_model(architecture, *layers, **hyperparameters):
    print(f"Compiling {architecture} with layers: {layers}")
    print(f"Settings: {hyperparameters}")

compile_model("VisionNet", "Conv2D", "MaxPool", "Dense", lr=0.001, epochs=50)
````

---


## What Happens Internally?

When Python sees this:

```python
@time_epoch
def compile_model(...):
```

it is similar to writing:

```python
compile_model = time_epoch(compile_model)
```

So when we call `compile_model()`, Python actually calls the `wrapper()` function first.

The wrapper records the start time, runs the original function, prints the execution time, and returns the original result.

This is a clean way to add logging, timing, authentication, or validation to functions without rewriting the original function body.

# Code Cell 4: Exception Handling

## Explanation

Real AI systems fail often. A dataset file may be missing, a student ID may not exist, a score may be outside the valid range, or a model checkpoint may fail to load.

Instead of letting the entire program crash, Python allows us to handle errors gracefully using exception handling.

This cell introduces custom exceptions, `try`, `except`, `raise`, and `finally`.

The custom exception `InvalidScoreError` makes the code easier to understand because the error name explains exactly what went wrong.

````python
class InvalidScoreError(Exception):
    pass

def log_assignment_score(student_id, score, max_score=150):
    try:
        if score < 0 or score > max_score:
            raise InvalidScoreError(f"{score} is invalid (Max {max_score}).")

        if student_id == "UNKNOWN":
            raise KeyError("Student ID not found in roster.")

        print(f"Logged {score}/{max_score} for {student_id}")

    except InvalidScoreError as e:
        print(f"Data Entry Error: {e}")
    except KeyError as e:
        print(f"Database Error: {e}")
    finally:
        print("Closing database connection...\n")

log_assignment_score("S-101", 140)
log_assignment_score("UNKNOWN", 98)
log_assignment_score("S-102", 200)
````

# Code Cell 5: File I/O and Context Managers

## Explanation

File I/O means reading from and writing to files.

In AI engineering, files are used for datasets, logs, model outputs, reports, configuration files, and experiment results.

This example writes a small conference draft into a text file and then reads it back.

The `with` statement is a context manager. It automatically closes the file after the block finishes. This is safer than manually opening and closing files.

````python
file_name = "DravidianLangTech_Draft.txt"
abstract = "Exploring morphological parsers for advanced NLP architectures."

with open(file_name, "w") as file:
    file.write("Title: Conference Submission 2026\n")
    file.write(f"Abstract: {abstract}")
    print("File written securely.")

with open(file_name, "r") as file:
    print("\n--- File Contents ---")
    print(file.read())
````

---


## Common File Modes

| Mode | Meaning |
|---|---|
| `"r"` | Read |
| `"w"` | Write and overwrite existing content |
| `"a"` | Append to the end of a file |
| `"x"` | Create a new file, fail if it already exists |

Important classroom note: using `"w"` will overwrite the file if it already exists.

# Code Cell 6: OOP Basics and State

## Explanation

Object-Oriented Programming is a way of organizing code using classes and objects.

A class is a blueprint. An object is an instance created from that blueprint.

In AI engineering, classes are used for models, datasets, data loaders, tokenizers, trainers, evaluators, and autograders.

This example creates an `Autograder` class.

The object remembers its internal state using `self.processed`. Each time we evaluate a batch, the processed count increases.

````python
class Autograder:
    def __init__(self, course, class_size):
        self.course = course
        self.class_size = class_size
        self.processed = 0

    def evaluate_batch(self, batch_size):
        self.processed += batch_size
        return f"[{self.course}] Processed {self.processed}/{self.class_size} students."

met_grader = Autograder("Spring 2026 AI", 123)

print(met_grader.evaluate_batch(10))
print(met_grader.evaluate_batch(40))
````

---


## Key OOP Terms

`class` defines the blueprint.

`__init__` runs automatically when an object is created.

`self` refers to the current object.

Instance variables store object-specific data:

```python
self.course
self.class_size
self.processed
```

The important idea is that the object remembers information between method calls.

# Code Cell 7: Encapsulation and Dunder Methods

## Explanation

Encapsulation means protecting internal data from direct external modification.

In this example, the neural layer has private weights using `self.__weights`.

The double underscore makes direct access difficult. This is useful when we want to protect important internal values such as neural network weights, configuration values, security tokens, or internal counters.

A dunder method has double underscores before and after its name.

The `__call__` method allows an object to behave like a function.

This pattern is common in deep learning frameworks. For example, PyTorch model objects can be called like functions during forward passes.

````python
class SecureNeuralLayer:
    def __init__(self, layer_name):
        self.layer_name = layer_name
        self.__weights = [0.25, 0.50, 0.88]

    def __call__(self, input_data):
        return f"Forward passing {input_data} through {self.layer_name}"

    def get_weights(self):
        return self.__weights

layer = SecureNeuralLayer("Hidden Layer 1")

print(layer("Image Tensor"))
print(f"Accessed safely: {layer.get_weights()}")
````

# Code Cell 8: Inheritance and Polymorphism

## Explanation

Inheritance allows one class to reuse and extend another class.

The parent class here is `BaseAIModel`.

The child classes are `VisionModel` and `NLPModel`.

Both child classes inherit from `BaseAIModel`.

Polymorphism means different objects can respond to the same method name in different ways.

Both child classes implement `process_data()`, but each one behaves differently.

This is extremely useful in AI pipelines where different model types need different processing logic.

````python
class BaseAIModel:
    def __init__(self, name):
        self.name = name

    def process_data(self):
        raise NotImplementedError("Child classes must implement this!")

class VisionModel(BaseAIModel):
    def process_data(self):
        return f"{self.name}: Processing 3D streetscape spatial simulations."

class NLPModel(BaseAIModel):
    def process_data(self):
        return f"{self.name}: Tokenizing language corpora."

pipeline_models = [VisionModel("ResNet"), NLPModel("BERT")]

for model in pipeline_models:
    print(model.process_data())
````

---


## Why This Matters in AI Engineering

Imagine a real AI system with different model types:

- Image classifier
- Speech recognizer
- Text classifier
- Object detector
- Recommendation model

Each model can have its own internal logic, but the pipeline can still call:

```python
model.process_data()
```

This makes the system clean, scalable, and easier to extend.

# Extra Teaching Example: Combining Functions, Files, and Classes

## Explanation

This mini-example combines Day 2 ideas into one small pipeline.

It creates a class called `ExperimentLogger`.

The logger stores an experiment name, writes results to a file, and uses a method to log new messages.

This is similar to how real machine learning experiments save logs.

````python
class ExperimentLogger:
    def __init__(self, experiment_name, file_name):
        self.experiment_name = experiment_name
        self.file_name = file_name

    def log(self, message):
        with open(self.file_name, "a") as file:
            file.write(f"[{self.experiment_name}] {message}\n")

    def show_logs(self):
        with open(self.file_name, "r") as file:
            print(file.read())

logger = ExperimentLogger("Day2_AI_Pipeline", "experiment_log.txt")

logger.log("Started experiment.")
logger.log("Loaded dataset successfully.")
logger.log("Model compiled successfully.")

logger.show_logs()
````
