## Slide 2: The Python Execution Model

### Explanation

Python code does not run directly as plain English-like text.

Python first compiles the code into bytecode. The Python Virtual Machine then executes that bytecode.

The `dis` module lets us inspect bytecode. This helps students understand that Python has an internal execution model, not just visible source code.

````python
import dis

def process_text(text):
    return text.lower().split().upper()

print("Disassembled Bytecode for process_text():")
dis.dis(process_text)
````

### Key Teaching Points

Teaching points:

- A function is defined using `def`.
- `text.lower()` converts text into lowercase.
- `.split()` breaks a string into tokens.
- `dis.dis()` shows bytecode instructions used internally by Python.

## Slide 3: PEP 8 & The Zen of Python

### Explanation

PEP 8 is the official style guide for Python. It helps us write code that is clean, readable, and professional.

In AI projects, readable code is very important because experiments are often updated many times.

````python
import this

textCorpus = "DravidianLangTech"

def count_chars(c):
    if len(c) > 0:
        return len(c)

text_corpus = "this is DravidianLangTech"

def count_characters(corpus):
    if corpus:
        return len(corpus)

print(count_chars("sai"))
print(count_characters(text_corpus))
````

## Slide 4: Variables and Memory References

### Explanation

Python variables are references to objects in memory.

When two variables point to the same list, changing one variable can affect the other. This matters in AI because large lists, arrays, tensors, images, and coordinate structures are often passed by reference.

````python
coord_a = []
coord_b = [10.5, 22.1, -5.0]

print(f"Memory ID of coord_a: {id(coord_a)}")
print(f"Memory ID of coord_b: {id(coord_b)}")

coord_a.append(1.0)
coord_a.append(2.0)

print(f"Value of coord_a after modification: {coord_a}")
print(f"Value of coord_b after modifying coord_a: {coord_b}")
````

### Key Teaching Points

Teaching points:

- `id()` shows the identity of an object in memory.
- `coord_b = coord_a` does not create a new list.
- Both variables refer to the same list object.
- Lists are mutable, so they can be changed after creation.

## Slide 6: Type Hinting

### Explanation

Type hints make code easier to understand, debug, and maintain.

## Slide 7: Advanced Operators

### Explanation

Python provides several operators that are useful for compact and efficient logic.

In this example:

- `//` performs floor division.
- `%` returns the remainder.
- `:=` is the walrus operator, which assigns and checks a value in the same expression.

````python
base_score = 100
bonus_points = 50

print(f"Batch processing groups of 10: {1234 // 100} full batches, {1234 % 10} remaining.")

if (total := base_score + bonus_points) == 140:
    print(f"Perfect bonus execution. Total score: {total}")
````

### Key Teaching Points

Teaching points:

- `123 // 10` gives the number of full groups.
- `123 % 10` gives the leftover amount.
- The walrus operator can reduce repeated calculations.
- Use the walrus operator only when it improves clarity.

## Slide 8: Truthiness and Falsiness

In Python, values can behave as either true or false inside conditions.

Falsy values include:

- Empty list `[]`
- Empty string `""`
- `None`
- `0`
- `False`

This helps us write clean checks for missing or empty data.

````python
empty_pixels = []
missing_text = ""
null_tensor = None

empty_pixels.append(1)

if not empty_pixels:
    empty_pixels.append(2)
    print("Warning: Pixel array is empty. Skipping convolution.")

if not missing_text:
    print("Warning: Text input required for NLP pipeline.")

if null_tensor is None:
    print("Warning: Tensor is missing.")

print(empty_pixels)
````

## Slide 9: Conditional Statements

### Explanation

Conditional statements allow a program to make decisions.

The `if/elif/else` structure checks conditions in order. Once one condition is true, Python runs that block and skips the remaining conditions.

The ternary operator is a shorter one-line version of simple conditional assignment.

````python
computer_vision_score = 99

if computer_vision_score >= 95:
    status = "Outstanding"
elif computer_vision_score >= 80:
    status = "Passing"
else:
    status = "Review Required"

deployment_ready = True if computer_vision_score == 98 else False

print(f"Status: {status} | Ready for Deployment: {deployment_ready}")
````

### Key Teaching Points

Teaching points:

- Conditions are checked from top to bottom.
- `elif` means “else if”.
- `else` catches all remaining cases.
- Ternary expressions are useful for simple assignments.

## Slide 10: Structural Pattern Matching

### Explanation

Structural pattern matching was introduced in Python 3.10.

It works similarly to switch-case statements in other languages. It is useful when we need to route different inputs to different processing pipelines.

In AI systems, this can help route text, image, audio, or numerical data.

````python
data_input = "Image"

match data_input:
    case "Text":
        print("Loading Dravidian language tokenizer...") 
    case "Image":
        print("Loading 2D Streetscape spatial map...")
    case "Audio":
        print("Loading spectrogram analyzer...")
    case _:
        print("Unknown modality. Defaulting to raw numerical array.")
````

### Key Teaching Points

Teaching points:

- `match` checks the value.
- Each `case` handles one possible pattern.
- `case _` is the default case.
- This is cleaner than long `if/elif` chains for many choices.

## Slide 11: The `for` Loop and Iterables

### Explanation

A `for` loop is used to iterate over a sequence.

Strings, lists, tuples, dictionaries, and ranges are iterable.

The `range()` function is memory-efficient because it does not store every number as a complete list. It produces numbers as needed.

````python
print("Tokenizing sequence:")
for char in "NLP":
    print(char.join("-"))

print("\nProcessing every 25th student in the cohort:")
for student_idx in range(1, 123,25):
    print(f"Evaluating student rank at index {student_idx}")
````

### Key Teaching Points

Teaching points:

- A string can be looped through character by character.
- `range(start, stop, step)` generates a sequence.
- The stop value is not included.
- Loops are fundamental for processing datasets.

## Slide 12: Supercharging the `for` Loop

### Explanation

Python provides tools that make loops more powerful.

`enumerate()` gives both the index and the value.

`zip()` allows us to loop through two or more sequences at the same time.

Together, they are useful for paired data such as IDs and scores, images and labels, or features and weights.

````python
student_ids = ["ID-01", "ID-02", "ID-03"]
cv_assignments = [98, 85, 92]

for index, (student, score) in enumerate(zip(student_ids, cv_assignments), start=1):
    print(f"Rank {index} | {student} scored {score}/98 on the CV task.")
````

### Key Teaching Points

Teaching points:

- `zip(student_ids, cv_assignments)` pairs values together.
- `enumerate(..., start=1)` gives rankings starting from 1.
- Tuple unpacking extracts each pair into `student` and `score`.
- This pattern is common in dataset inspection and reporting.

## Slide 13: The `while` Loop

### Explanation

A `while` loop runs as long as a condition remains true.

Unlike a `for` loop, we do not always know exactly how many times a `while` loop will run.

This is useful in training processes where a loop runs until a score, loss, threshold, or convergence condition is reached.

````python
current_score = 120
max_score = 150
bonus_awarded = 0

print("Processing grading appeals...")

while current_score < 140:
    current_score += 10
    bonus_awarded += 10
    print(f"Adjusting score... Current: {current_score}")

print(f"Final Score: {current_score}/{max_score}. Total bonus applied: {bonus_awarded}")
````

### Key Teaching Points

Teaching points:

- A `while` loop needs a condition.
- The condition must eventually become false.
- If the condition never changes, an infinite loop may occur.
- Update variables carefully inside the loop.

## Slide 14: Loop Control

### Explanation

Python gives us special keywords to control loop behavior.

`break` exits a loop completely.

`continue` skips the current iteration and moves to the next iteration.

A loop `else` block runs only if the loop finishes normally without hitting `break`.

````python
simulated_files = ["street_1.obj", "street_2.obj", "street_3.obj"]
corrupt_flag = "error"

for file in simulated_files:
    if corrupt_flag in file:
        print(f"Corrupted mesh found in {file}. Halting pipeline.")
        break
    print(f"Successfully rendered {file}")
else:
    print("All 3D streetscape meshes loaded successfully without errors.")
````

### Key Teaching Points

Teaching points:

- `break` stops the loop immediately.
- `else` after a loop is unique and useful for search-style tasks.
- The `else` block means “no break happened”.
- This pattern can be used for validation pipelines.

## Slide 15: Day 1 Wrap-up Mini-Capstone

### Explanation

This final example combines several Python ideas into one small preprocessing pipeline.

It uses:

- Function definition
- Type hints
- Truthiness
- List comprehension
- String cleaning
- Looping over cleaned output

This is similar to preprocessing raw data before sending it into an AI model.

````python
def process_course_materials(materials: list) -> list:
    """Filters out empty submissions and normalizes text."""
    if not materials:
        return []

    cleaned = [text.strip().lower() for text in materials if text and text.strip()]
    return cleaned

raw_submissions = ["  Dravidian Morphological Parser  ", "", "3D Spatial Analysis", None, "  "]
valid_data = process_course_materials(raw_submissions)

print("Cleaned Dataset for Model Input:")
for item in valid_data:
    print(f"- {item}")
````

### Key Teaching Points

Teaching points:

- `if not materials:` handles empty input.
- List comprehensions can combine looping, filtering, and transformation.
- `.strip()` removes unwanted spaces.
- `.lower()` normalizes text.
- Clean data is essential before training or testing AI models.
